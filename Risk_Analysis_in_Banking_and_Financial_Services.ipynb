```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_excel("/content/Banking.xlsx")
```

**Generating descriptive statistics for the dataframe**

```python
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['low', 'Medium', 'high']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

**Value Counts of Categorical Columns**

```python
categorical_cols=df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]

for col in categorical_cols:
  print(f"Value count for '{col}': ")
  display(df[col].value_counts())
```

**Univariate Analysis**

```python
for i, predictor in enumerate(df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

**Bivariate Analysis**

```python
for i, predictor in enumerate(df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor,hue='Nationality')
```

**Numerical Analysis**

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#Univariate analysis and visualization

plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
plt.tight_layout()
plt.show()
```

**HeatMap**

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,8))

sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.show()
```
